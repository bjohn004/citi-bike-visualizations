```python
import pandas as pd
```

```python
csv2017 = "Resources/201712-citibike-tripdata.csv"
csv2018 = "Resources/201812-citibike-tripdata.csv"
csv2019 = "Resources/201912-citibike-tripdata.csv"
csv2020 = "Resources/202012-citibike-tripdata.csv"
csv2021 = "Resources/202112-citibike-tripdata.csv"
csv2022 = "Resources/202209-citibike-tripdata.csv"

df_2017 = pd.read_csv(csv2017)
df_2018 = pd.read_csv(csv2018)
df_2019 = pd.read_csv(csv2019)
df_2020 = pd.read_csv(csv2020)
df_2021 = pd.read_csv(csv2021)
df_2022 = pd.read_csv(csv2022)
```

```python
df_2017["usertype"] = df_2017["usertype"].replace("Subscriber","member")
df_2017["usertype"] = df_2017["usertype"].replace("Customer","casual")
df_2017["year"]="2017"
df_2018["usertype"] = df_2018["usertype"].replace("Subscriber","member")
df_2018["usertype"] = df_2018["usertype"].replace("Customer","casual")
df_2018["year"]="2018"
df_2019["usertype"] = df_2019["usertype"].replace("Subscriber","member")
df_2019["usertype"] = df_2019["usertype"].replace("Customer","casual")
df_2019["year"]="2019"
df_2020["usertype"] = df_2020["usertype"].replace("Subscriber","member")
df_2020["usertype"] = df_2020["usertype"].replace("Customer","casual")
df_2020["year"]="2020"
```

```python
df_2017 = df_2017.drop("gender", axis=1)
df_2017 = df_2017.drop("birth year", axis=1)
df_2018 = df_2018.drop("gender", axis=1)
df_2018 = df_2018.drop("birth year", axis=1)
df_2019 = df_2019.drop("gender", axis=1)
df_2019 = df_2019.drop("birth year", axis=1)
df_2020 = df_2020.drop("gender", axis=1)
df_2020 = df_2020.drop("birth year", axis=1)
```

```python
df_2017 = df_2017.rename(columns={"bikeid": "ride_id", "starttime":"started_at","stoptime": "ended_at", "start station id": "start_station_id", "start station name": "start_station_name", "start station latitude":"start_lat", "start station longitude": "start_lng", "end station id": "end_station_id", "end station name": "end_station_name", "end station latitude":"end_lat", "end station longitude": "end_lng", "usertype":"member_casual"})
df_2018 = df_2018.rename(columns={"bikeid": "ride_id", "starttime":"started_at","stoptime": "ended_at", "start station id": "start_station_id", "start station name": "start_station_name", "start station latitude":"start_lat", "start station longitude": "start_lng", "end station id": "end_station_id", "end station name": "end_station_name", "end station latitude":"end_lat", "end station longitude": "end_lng", "usertype":"member_casual"})
df_2019 = df_2019.rename(columns={"bikeid": "ride_id", "starttime":"started_at","stoptime": "ended_at", "start station id": "start_station_id", "start station name": "start_station_name", "start station latitude":"start_lat", "start station longitude": "start_lng", "end station id": "end_station_id", "end station name": "end_station_name", "end station latitude":"end_lat", "end station longitude": "end_lng", "usertype":"member_casual"})
df_2020 = df_2020.rename(columns={"bikeid": "ride_id", "starttime":"started_at","stoptime": "ended_at", "start station id": "start_station_id", "start station name": "start_station_name", "start station latitude":"start_lat", "start station longitude": "start_lng", "end station id": "end_station_id", "end station name": "end_station_name", "end station latitude":"end_lat", "end station longitude": "end_lng", "usertype":"member_casual"})
```

```python
df_2021=df_2021.drop("rideable_type",axis=1)
df_2022=df_2022.drop("rideable_type",axis=1)
df_2021["year"]="2021"
df_2022["year"]="2022"
```

```python
df_2017_2022 = pd.concat([df_2017, df_2018, df_2019, df_2020, df_2021, df_2022], ignore_index=True)
```

```python
df_2017_2022["City"] = "New York City"
```
